```python
import matplotlib.pyplot as plt 
import pandas 
import warnings 
import seaborn
from sklearn.cluster import KMeans
warnings.filterwarnings('ignore')
```

```python
k = pandas.read_csv('D:\Mall_Customers.csv', index_col = 'CustomerID')
k.head()
```

```python
k.describe()
```

```python
k.isnull().sum()
```

```python
k.drop_duplicates(inplace=True)
```

```python
k.isnull()
```

```python
x = k.iloc[:,[2,3]].values
print(x)
```

```python
wcss = []
for i in range (1,11):
    kmeans = KMeans(n_clusters =i, init ='k-means++', random_state = 42)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
```

```python
plt.figure(figsize=(10,5))
seaborn.lineplot(range(1,11),wcss,color = 'blue')
plt.xlabel('no. of clusters')
plt.ylabel('wcss')
plt.show()
```

```python
seaborn.violinplot('Spending_Score','Genre',data=k,palette='coolwarm')
```

```python
kmeans = KMeans(n_clusters =5, init ='k-means++', random_state = 42)
y_pred = kmeans.fit_predict(x)
plt.figure(figsize=(10,6))
for i in range (5):
    plt.scatter(x[y_pred == i,0],x[y_pred == i,1])
  
```
